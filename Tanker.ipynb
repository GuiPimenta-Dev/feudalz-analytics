```python
df = simulation(my_defense_bonus=13.8,my_attack_bonus=0,enemy_defense_bonus=13.8,hero="double",rarity="epic", group=True)
df1 = simulation(my_defense_bonus=13.8,my_attack_bonus=0,enemy_defense_bonus=13.8, group=True)
```
